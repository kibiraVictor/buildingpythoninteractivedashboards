```python
import plotly.express as px
```

```python
gapminder = px.data.gapminder()
```

```python
gapminder
```

```python
px.scatter(data_frame=gapminder,
          x='gdpPercap',
          y='lifeExp',
          size='pop',
          facet_col='continent',
          color='continent',
          title='Life Expectancy and GDP per CAPITA.  1952-2007',
          labels={
              'gdpPercap': 'GDP per Capita',
              'lifeExpr': 'Life Expectancy'},
          log_x = True,
          range_y = [20, 100],
          hover_name ='country',
          animation_frame='year',
          height=600,
          size_max=90)
```
