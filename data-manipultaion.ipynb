```python
data = pd.read_csv('data/PovStatsData.csv')
```

```python
data
```

```python
data.columns
```

```python
data = data.drop('Unnamed: 50', axis=1)
```

```python
data.shape
```

```python
data.sample(3)
```

```python
data.head()
```

```python
data.loc[:].isna().mean().mean()
```

```python
#melting our data dataframe
```

```python
data.columns
```

```python
#check on data melting and data pivoting
```

```python
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
```

```python
data_melt = data.melt(id_vars=id_vars, var_name='year').dropna(subset=['value'])
```

```python
data_melt
```

```python
data_melt['year'] = data_melt['year'].astype(int)
```

```python
print(data_melt.shape)
```

```python
data_melt.loc[:].isna().mean().mean()
```

```python
#pivoting our data;kinda going back to the original place before melting our data
```

```python
data_melt.columns
```

```python
data_pivot =\
data_melt.pivot(index=['Country Name', 'Country Code', 'year'], columns='Indicator Name', values='value').reset_index()
```

```python
data_pivot.shape
```

```python
data_pivot.sample(5)
```

```python
poverty = pd.merge(data_pivot, country,
                  left_on='Country Code',
                  right_on='Country Code',
                  how='left')
```

```python
poverty.shape
```

```python
poverty
```

```python
poverty[['Country Code', 'year']].duplicated().any()
```

```python
poverty.to_csv('povertydata.csv', index=False)
```
